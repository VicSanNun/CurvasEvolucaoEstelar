```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class ExtractData:
    def extractLogL_And_LogTeff(self, file):
        
        logL = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            logL.append(lines[index].split()[6])
            logTeff.append(lines[index].split()[11])
            
        #Adicionando as informações em um np.Array
        logLNp = np.array(logL)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatlogLNp = logLNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatlogLNp, floatlogTeffNp]
    
    def extractLogG_And_LogTeff(self, file):
        
        logG = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            logG.append(lines[index].split()[16])
            logTeff.append(lines[index].split()[13])
            
        #Adicionando as informações em um np.Array
        logGNp = np.array(logG)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatlogGNp = logGNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatlogGNp, floatlogTeffNp]
        
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

exData = ExtractData()

Mass15 = exData.extractLogL_And_LogTeff('EvoTrack15.txt')
Mass20 = exData.extractLogL_And_LogTeff('EvoTrack20.txt')
Mass25 = exData.extractLogL_And_LogTeff('EvoTrack25.txt')
Mass30 = exData.extractLogL_And_LogTeff('EvoTrack30.txt')
Mass40 = exData.extractLogL_And_LogTeff('EvoTrack40.txt')

_01e9Year = exData.extractLogG_And_LogTeff('iso_01.txt')


#Plotando
fig, ax = plt.subplots()
ax.plot(Mass15[1], Mass15[0], label=r'$15 M_{\odot}$')
ax.plot(Mass20[1], Mass20[0], label=r'$20 M_{\odot}$')
ax.plot(Mass25[1], Mass25[0], label=r'$25 M_{\odot}$')
ax.plot(Mass30[1], Mass30[0], label=r'$30 M_{\odot}$')
ax.plot(Mass40[1], Mass40[0], label=r'$40 M_{\odot}$')

ax.set(xlabel=r'$Log(T_{eff})$', ylabel=r'$Log(L)$')

plt.xlim(4.8,3.9)
plt.ylim(4.2,6)

plt.legend()

if (os.path.exists("questao_1.png") == False):
    fig.savefig("questao_1")

plt.show()
```

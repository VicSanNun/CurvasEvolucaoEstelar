# Classe com os Métodos de Extração de Dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class ExtractData:
    def extractLogL_And_LogTeff(self, file):
        
        logL = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            logL.append(lines[index].split()[6])
            logTeff.append(lines[index].split()[11])
            
        #Adicionando as informações em um np.Array
        logLNp = np.array(logL)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatlogLNp = logLNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatlogLNp, floatlogTeffNp]
    
    def extractLogG_And_LogTeff(self, file):
        #Para Isócronas
        logG = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            logG.append(lines[index].split()[16])
            logTeff.append(lines[index].split()[13])
            
        #Adicionando as informações em um np.Array
        logGNp = np.array(logG)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatlogGNp = logGNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatlogGNp, floatlogTeffNp]
    
    def extractStarAge_And_Radius(self, file):
        
        StarAge = []
        logR = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            logR.append(lines[index].split()[13])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        logRNp = np.array(logR)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatlogRNp = logRNp.astype(float)
        
        return [floatStarAgeNp, floatlogRNp]
    
    def extractStarAge_And_LogTeff(self, file):
        
        StarAge = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            logTeff.append(lines[index].split()[11])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatStarAgeNp, floatlogTeffNp]
    
    def extractStarAge_And_LogTeff(self, file):
        
        StarAge = []
        logTeff = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            logTeff.append(lines[index].split()[11])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        logTeffNp = np.array(logTeff)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatlogTeffNp = logTeffNp.astype(float)
        
        return [floatStarAgeNp, floatlogTeffNp]
    
    def extractStarAge_And_CoreEnergyGenerate(self, file):
        
        StarAge = []
        pp = []
        cno = []
        tri_alpha = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            pp.append(lines[index].split()[6])
            cno.append(lines[index].split()[60])
            tri_alpha.append(lines[index].split()[61])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        ppNp = np.array(pp)
        cnoNp = np.array(cno)
        triAlphaNp = np.array(tri_alpha)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatppNp = ppNp.astype(float)
        floatcnoNp = cnoNp.astype(float)
        floattriAlphaNp = triAlphaNp.astype(float)
        
        return [floatStarAgeNp, floatppNp, floatcnoNp, floattriAlphaNp]
    
    def extractStarAge_And_CoreTemp(self, file):
        
        StarAge = []
        CoreTemp = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            CoreTemp.append(lines[index].split()[45])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        CoreTempNp = np.array(CoreTemp)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatCoreTempNp = CoreTempNp.astype(float)
        
        return [floatStarAgeNp, floatCoreTempNp]
    
    def extractStarAge_And_CoreDensity(self, file):
        
        StarAge = []
        CoreDensity = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            CoreDensity.append(lines[index].split()[46])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        CoreDensityNp = np.array(CoreDensity)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatCoreDensityNp = CoreDensityNp.astype(float)
        
        return [floatStarAgeNp, floatCoreDensityNp]
    
    def extractStarAge_And_StarMass(self, file):
        #Na verdade extrai o Log(HBurning)
        StarAge = []
        StarMass = []
        
        with open(file) as f:
            lines = f.readlines()
        
        for index, linha in enumerate(lines):
            StarAge.append(lines[index].split()[0])
            StarMass.append(lines[index].split()[52])
            
        #Adicionando as informações em um np.Array
        StarAgeNp = np.array(StarAge)
        StarMassNp = np.array(StarMass)

        #Convertendo os dados para Float
        floatStarAgeNp = StarAgeNp.astype(float)
        floatStarMassNp = StarMassNp.astype(float)
        
        return [floatStarAgeNp, floatStarMassNp]
        
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

exData = ExtractData()

Mass15 = exData.extractLogL_And_LogTeff('EvoTrack15.txt')
Mass20 = exData.extractLogL_And_LogTeff('EvoTrack20.txt')
Mass25 = exData.extractLogL_And_LogTeff('EvoTrack25.txt')
Mass30 = exData.extractLogL_And_LogTeff('EvoTrack30.txt')
Mass40 = exData.extractLogL_And_LogTeff('EvoTrack40.txt')

_01e9Year = exData.extractLogG_And_LogTeff('iso_01.txt')


#Plotando
fig, ax = plt.subplots()
ax.plot(Mass15[1], Mass15[0], label=r'$15 M_{\odot}$')
ax.plot(Mass20[1], Mass20[0], label=r'$20 M_{\odot}$')
ax.plot(Mass25[1], Mass25[0], label=r'$25 M_{\odot}$')
ax.plot(Mass30[1], Mass30[0], label=r'$30 M_{\odot}$')
ax.plot(Mass40[1], Mass40[0], label=r'$40 M_{\odot}$')

ax.set(xlabel=r'$Log(T_{eff})$', ylabel=r'$Log(L)$', title="Questão 1")

plt.xlim(4.8,3.9)
plt.ylim(4.2,6)

plt.legend()

# if (os.path.exists("questao_1.png") == False):
#     fig.savefig("questao_1")

fig.savefig("questao_1")

plt.show()
```
